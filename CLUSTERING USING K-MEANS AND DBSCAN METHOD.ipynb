```python
#Normalize data to have similar scale using StandardScalar function to fit the selected variables
from sklearn.preprocessing import StandardScaler
X =  Country_data.iloc[:, [1,3]].values
sc_X= StandardScaler()
X= sc_X.fit_transform(X)
```

```python
#Normalise data without outlier in child mortality
X2 =  New_Country_data.iloc[:, [1,3]].values
sc_X= StandardScaler()
X2= sc_X.fit_transform(X2)
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss= []
for i in range(1,8):
    kmeans= KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,8), wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#find the optimal value of k without outlier
from sklearn.cluster import KMeans
wcss2= []
for i in range(1,9):
    kmeans2= KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans2.fit(X2)
    wcss2.append(kmeans2.inertia_)
plt.plot(range(1,9), wcss2)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS2')
plt.show()
```

```python
#fitting k-Means to the dataset with the identified number of clusters
kmeans= KMeans(n_clusters = 3, init = 'k-means++', random_state= 42)
# predict y values  using the model
y_kmeans = kmeans.fit_predict(X)
```

```python
# repeat for X2
kmeans2= KMeans(n_clusters = 3, init = 'k-means++', random_state= 42)

y_kmeans2 = kmeans2.fit_predict(X2)
```

```python
#visualize the clusters derived by the model
colours=['red','blue','green','cyan','black']
plt.figure(figsize=(8,8))

for i in range(3):
    plt.scatter(X[y_kmeans== i,0],X[y_kmeans== i,1], s=100, c=colours[i],label= 'Cluster' + str(i+1))
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s= 100, c= 'yellow', label= 'Centroids')
plt.title('Cluster of Countries')
plt.xlabel('Child Mortality Rate(Scaled)')
plt.ylabel('Health Spending(% GDP)(Scaled)')
plt.legend()
plt.show()
```

```python
#visualize cluster for dataset without outlier 
colours=['red','blue','green','cyan','black']
plt.figure(figsize=(8,8))

for i in range(3):
    plt.scatter(X2[y_kmeans2== i,0],X2[y_kmeans2== i,1], s=100, c=colours[i],label= 'Cluster' + str(i+1))
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1], s= 100, c= 'yellow', label= 'Centroids')
plt.title('Cluster of Countries with Removed Maximum Child Mortality')
plt.xlabel('Child Mortality Rate(Scaled)')
plt.ylabel('Health Spending(% GDP)(Scaled)')
plt.legend()
plt.show()
```

```python
# Density clustering using DBSCAN
# determine Epsilon Value using KNN and sort distance in ascending order
from sklearn.neighbors import NearestNeighbors
neighbours= NearestNeighbors(n_neighbors = 3)
distances, indices= neighbours.fit(X).kneighbors(X)
distances= distances[:,1]
distances=np.sort(distances,axis=0)
plt.title("Determine Epsilon Value using KNN vs Distances")
plt.plot(distances)
```

```python
# repeat for X2
from sklearn.neighbors import NearestNeighbors
neighbours= NearestNeighbors(n_neighbors = 3)
distances, indices= neighbours.fit(X2).kneighbors(X2)
distances= distances[:,1]
distances=np.sort(distances,axis=0)
plt.plot(distances)
```

```python
#instantiate DBSCAN model using the identified Epsilon Value and minimum sample of 3 to predict y values
from sklearn.cluster import DBSCAN
dbscan= DBSCAN(eps=0.30, min_samples= 4)
y_dbscan= dbscan.fit_predict(X)
```

```python
#inspect the array to identify the number of clusters
y_dbscan
```

```python
#print identified cluster labels from the model
print ("Clusters assigned are:", set(dbscan.labels_))
```

```python
#visualize the clusters derived by the model
plt.figure(figsize=(8,8))
colours=['yellow','blue','green','orange','black','purple','cyan']

for i in range(4):
    plt.scatter(X[y_dbscan== i,0],X[y_dbscan== i,1], s=100, c=colours[i],label= 'Cluster' + str(i+1))
plt.scatter(X[y_dbscan== -1,0],X[y_dbscan== -1,1], s= 100, c= 'red', label = 'Noise')
plt.title('Cluster of Countries')
plt.xlabel('Child Mortality Rate(Scaled)')
plt.ylabel('Health Spending(% GDP)(Scaled)')
plt.legend()
plt.show()
```

```python
#instantiate DBSCAN model using the identified Epsilon Value and minimum sample of 4 to predict y values without outlier
from sklearn.cluster import DBSCAN
dbscan2= DBSCAN(eps=0.30, min_samples= 4)
y_dbscan2= dbscan2.fit_predict(X2)
```

```python
#print identified cluster labels from the model
print ("Clusters assigned are:", set(dbscan2.labels_))
```

```python
#plot the clusters derived by the model
plt.figure(figsize=(8,8))
colours=['yellow','blue','green','orange','black','purple','cyan']

for i in range(4):
    plt.scatter(X2[y_dbscan2== i,0],X2[y_dbscan2== i,1], s=100, c=colours[i],label= 'Cluster' + str(i+1))
plt.scatter(X2[y_dbscan2== -1,0],X2[y_dbscan2== -1,1], s= 100, c= 'red', label = 'Noise')
plt.title('Cluster of Countries')
plt.xlabel('Child Mortality Rate(Scaled)')
plt.ylabel('Health Spending(% GDP)(Scaled)')
plt.legend()
plt.show()
```

```python
# Evaluate both model silhoutte scores for quality
from sklearn.metrics import silhouette_score
kmeans_silhouette = silhouette_score(X, kmeans.labels_)
kmeans_silhouette2 = silhouette_score(X2, kmeans2.labels_)
dbscan_silhouette = silhouette_score(X, dbscan.labels_)
dbscan_silhouette2 = silhouette_score(X2, dbscan2.labels_)
```
